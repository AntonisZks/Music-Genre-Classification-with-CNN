## ŒöŒ±œÑŒ≠Œ≤Œ±œÉŒºŒ± ŒúŒøœÖœÉŒπŒ∫ŒÆœÇ Œ±œÄœå œÑŒø YouTube üé•‚ñ∂Ô∏èüì©

ŒìŒπŒ± ŒΩŒ± Œ∫Œ±œÑŒµŒ≤Œ¨œÉŒøœÖŒºŒµ ŒºŒøœÖœÉŒπŒ∫ŒÆ Œ±œÄœå œÑŒø Œ¥ŒπŒ±Œ¥ŒØŒ∫œÑœÖŒø Œ∏Œ± œáœÅŒ∑œÉŒπŒºŒøœÄŒøŒπŒÆœÉŒøœÖŒºŒµ Œ≠ŒΩŒ± module ŒøŒΩœåŒºŒ±œÑŒπ [pytube][pytube_link]. ŒëœÇ Œ¥Œ∑ŒºŒπŒøœÖœÅŒ≥ŒÆœÉŒøœÖŒºŒµ ŒºŒØŒ± __œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑__ `download_music()` Œ∑ ŒøœÄŒøŒØŒ± Œ¥ŒøœÉŒºŒ≠ŒΩŒøœÖ ŒµŒΩœåœÇ _youtube url_ Œ∫Œ±œÑŒµŒ≤Œ¨Œ∂ŒµŒπ œÑŒø Œ±ŒΩœÑŒØœÉœÑŒøŒπœáŒø Œ≤ŒØŒΩœÑŒµŒø Œ∫Œ±Œπ œÑŒø Œ±œÄŒøŒ∏Œ∑Œ∫Œ≠œÖŒµŒπ œÉŒµ ŒºŒøœÅœÜŒÆ __.mp3__ œÉœÑŒø œÜŒ¨Œ∫ŒµŒªŒø `/Data/Music/`

[pytube_link]: https://pytube.io/en/latest/

```python
from pytube import YouTube
import os

def download_music(url: str, category: str) -> bool:
    """ Downloads a youtube video from YouTube according to its given url, and saves it 
        in a .mp3 format """

    with open(resource_path(f'../Data/Music/downloaded.txt'), 'r') as fp:
        lines = fp.readlines()
        for downloaded_url in lines:
            if downloaded_url == url:
                return False
    
    # Create a YouTube object and pass the given video url, and extract only its audio
    yt = YouTube(url)
    video = yt.streams.filter(only_audio=True).first()

    # Download the mp3 file in an appropriate destination
    out_file = video.download(output_path=resource_path(f"..\\Data\\Music\\{category}"))

    # Save the file
    base, ext = os.path.splitext(out_file) 
    new_file = base + '.mp3'
    os.rename(out_file, new_file)

    with open(resource_path(f'../Data/Music/downloaded.txt'), 'a') as fp:
        fp.write(url)

    return True
```

```python
import os

# Define a list with the categories we have
CATEGORIES_NAMES = ['blues', 'classical', 'hiphop', 'rock']
CATEGORIES_CODES = ['0', '1', '2', '3']

# Loop through every url of every category and download its video from youtube abd save it as an mp3 format
for category in CATEGORIES_NAMES:
    with open(resource_path(f'../Data/Songs_urls/{category}.txt'), 'r') as fp:
        for url in fp.readlines():
            download_music(url, category)

print("Download Process Done!")
```

ŒàœáŒøŒΩœÑŒ±œÇ Œ∫Œ±œÑŒµŒ≤Œ¨œÉŒµŒπ ŒµœÄŒπœÑœÖœáœéœÇ œÑŒ± œÑœÅŒ±Œ≥ŒøœçŒ¥ŒπŒ± ŒºŒ±œÇ Œ∏Œ± œÑŒ± __ŒºŒµœÑŒ±œÉœáŒ∑ŒºŒ±œÑŒØœÉŒøœÖŒºŒµ__ œÉŒµ Œ∫Œ±œÑŒ¨ŒªŒªŒ∑ŒªŒ∑ __ŒºŒøœÅœÜŒÆ__ Œ≥ŒπŒ± œÑŒ∑ŒΩ ŒµŒØœÉŒøŒ¥Œø œÑŒøœÖ ŒºŒøŒΩœÑŒ≠ŒªŒøœÖ ŒºŒ±œÇ. ŒëœÖœÑŒÆ Œ∑ ŒºŒøœÅœÜŒÆ ŒªŒ≠Œ≥ŒµœÑŒ±Œπ __[œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ¨œÜŒ∑ŒºŒ±][spectrograms_link]__. ŒìŒπŒ± Œ±œÅœáŒÆ Œ∏Œ± Œ±œÄŒøŒ∏Œ∑Œ∫ŒµœçœÉŒøœÖŒºŒµ œÑŒ± Œ∑œáŒ∑œÑŒπŒ∫Œ¨ ŒºŒ±œÇ Œ±œÅœáŒµŒØŒ± œÉŒµ ŒºŒøœÅœÜŒÆ [audio waveform][audio_waveform_link], Œ≠œÑœÉŒπ œéœÉœÑŒµ ŒΩŒ± ŒºœÄŒøœÅŒøœçŒºŒµ Œ±œÅŒ≥œåœÑŒµœÅŒ± ŒΩŒ± ŒµŒæŒ¨Œ≥ŒøœÖŒºŒµ œÑŒø œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ¨œÜŒ∑ŒºŒ± œÑŒøœÖ Œ∫Œ±Œ∏ŒµŒΩœåœÇ. ŒìŒπ' Œ±œÖœÑœå Œ∏Œ± œáœÅŒ∑œÉŒπŒºŒøœÄŒøŒπŒÆœÉŒøœÖŒºŒµ œÑŒø œÄŒ±Œ∫Œ≠œÑŒø [librosa][librosa_link] œÑŒ∑œÇ Python.

[librosa_link]: https://librosa.org/doc/latest/index.html
[audio_waveform_link]: https://en.wikipedia.org/wiki/Waveform
[spectrograms_link]: https://en.wikipedia.org/wiki/Spectrogram#:~:text=A%20spectrogram%20is%20a%20visual,sonographs%2C%20voiceprints%2C%20or%20voicegrams

```python
import warnings
from glob import glob
import librosa
import librosa.display

warnings.filterwarnings('ignore') # Just to ignore some warnings below

# Create a dictionary that will keep all the audio waveforms of each category
music_audio = {key: [] for key in CATEGORIES_NAMES}

# Loop through every audio in every category and store its waveform
for category in CATEGORIES_NAMES:
    audio_files = glob(resource_path(f'../Data/Music/{category}/*.mp3'))

    # Extract the waveform of the audios
    for audio_file in audio_files:
        audio_samples, sampling_rate = librosa.load(audio_file)
        duration = librosa.get_duration(y=audio_samples, sr=sampling_rate)
        music_audio[category].append((audio_samples, sampling_rate, audio_file, duration))
```

Œ£œÖŒ≥Œ∫ŒµŒ∫œÅŒπŒºŒ≠ŒΩŒ± Œ∏Œ± Œ¥Œ∑ŒºŒπŒøœÖœÅŒ≥ŒÆœÉŒøœÖŒºŒµ Œ≠ŒΩŒ± Œ¥ŒπŒ¨Œ≥œÅŒ±ŒºŒºŒ± ŒºŒµ __4 waveforms__, œÑŒø œÄœÅœéœÑŒø Œ≥ŒπŒ± Œ∫Œ¨Œ∏Œµ Œ∫Œ±œÑŒ∑Œ≥ŒøœÅŒØŒ± ŒºŒøœÖœÉŒπŒ∫ŒÆœÇ.

```python
from matplotlib import pyplot as plt
%matplotlib inline

SAMPLE_TO_SHOW = 0 # The first sample

style = 'default' # 'dark_background'
plt.style.use(style)

fig, ax = plt.subplots(4, 1, figsize=(14, 11))

# Plot the waveforms
for i, category in enumerate(CATEGORIES_NAMES):
    y = music_audio[category][SAMPLE_TO_SHOW][0]
    ax[i].plot(range(len(y)), y, linewidth=0.2, color=custom_hex_color('#4194FF'))
    ax[i].set_title("Category: " + category + " \n Song: " + music_audio[category][0][2].split('\\')[-1][:-4])
    ax[i].set_xlabel('Time (s)')
    ax[i].set_ylabel('Amplitude')

fig.suptitle('\nFirst song waveform of every music category\n', fontsize=25)
plt.tight_layout()
plt.show()
```

## ŒïŒæŒ±Œ≥œâŒ≥ŒÆ Œ¶Œ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒ∑ŒºŒ¨œÑœâŒΩ ŒºŒµ œáœÅŒÆœÉŒ∑ œÑŒøœÖ ŒºŒµœÑŒ±œÉœáŒ∑ŒºŒ±œÑŒπœÉŒºŒøœç Fourier üìà‚úçüèªüß™

Œ£œÑŒ∑ œÉœÖŒΩŒ≠œáŒµŒπŒ± Œ∏Œ± ŒµŒæŒ¨Œ≥ŒøœÖŒºŒµ œÑŒ± __œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒÆŒºŒ±œÑŒ±__ œÑœâŒΩ waveforms œÄŒøœÖ ŒºœåŒªŒπœÇ ŒªŒ¨Œ≤Œ±ŒºŒµ. ŒìŒπŒ± œÑŒøŒΩ ŒªœåŒ≥Œø Œ±œÖœÑœå Œ∏Œ± ŒºŒµœÑŒ±Œ≤ŒøœçŒºŒµ Œ±œÄœå œÑŒø œÄŒµŒ¥ŒØŒø œÑŒøœÖ __œáœÅœåŒΩŒøœÖ__, œÉœÑŒø œÄŒµŒ¥ŒØŒø œÑŒ∑œÇ __œÉœÖœáŒΩœåœÑŒ∑œÑŒ±œÇ__, œáœÅŒ∑œÉŒπŒºŒøœÄŒøŒπœéŒΩœÑŒ±œÇ œÑŒøŒΩ [ŒºŒµœÑŒ±œÉœáŒ∑ŒºŒ±œÑŒπœÉŒºœå fourier][fourier_link] (Fourier Transformation). Œ£œÖŒ≥Œ∫ŒµŒ∫œÅŒπŒºŒ≠ŒΩŒ± Œ∏Œ± œáœÅŒ∑œÉŒπŒºŒøœÄŒøŒπŒÆœÉŒøœÖŒºŒµ œÑŒøŒΩ [Œ≥œÅŒÆŒ≥ŒøœÅŒø ŒºŒµœÑŒ±œÉœáŒ∑ŒºŒ±œÑŒπœÉŒºœå fourier][short_fourier_link] Œ≥ŒπŒ± __Œ¥ŒπŒ±Œ∫œÅŒπœÑŒ≠œÇ œÑŒπŒºŒ≠œÇ__ œÉŒÆŒºŒ±œÑŒøœÇ Œ≥ŒπŒ± œÑŒøŒΩ ŒøœÄŒøŒØŒ± œÖœÄŒ¨œÅœáŒµŒπ Œ≠œÑŒøŒπŒºŒ∑ œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑ œÉœÑŒ∑ Œ≤ŒπŒ≤ŒªŒπŒøŒ∏ŒÆŒ∫Œ∑ __librosa__. Œ£œÑŒ∑ Œ¥ŒπŒ∫ŒπŒ¨ ŒºŒ±œÇ œÄŒµœÅŒØœÄœÑœâœÉŒ∑ œÑŒø _œÉŒÆŒºŒ±_ ŒµŒØŒΩŒ±Œπ œÑŒø __waveform__ œÑŒøœÖ Œ∫Œ¨Œ∏Œµ œÑœÅŒ±Œ≥ŒøœÖŒ¥ŒπŒøœç Œ∫Œ±Œπ Œø œÑœçœÄŒøœÇ œÑŒøœÖ __Œ¥ŒπŒ±Œ∫œÅŒπœÑŒøœç ŒºŒµœÑŒ±œÉœáŒ∑ŒºŒ±œÑŒπœÉŒºŒøœç fourier__ ŒµŒØŒΩŒ±Œπ Œø Œ±Œ∫œåŒªŒøœÖŒ∏ŒøœÇ:

$$
    \boxed{X(m, \omega) = \sum_{n=-\infty}^{\infty} x_n w_{n-m}e^{-j\omega n}}
$$

[fourier_link]: https://en.wikipedia.org/wiki/Fourier_transform
[short_fourier_link]: https://en.wikipedia.org/wiki/Short-time_Fourier_transform

```python
import librosa
import numpy as np

music_spectograms = {key: [] for key in CATEGORIES_NAMES}

# Compute the spectrograms for every song of every category
for category in CATEGORIES_NAMES:
    for i in range(len(music_audio[category])):
        D = librosa.stft(music_audio[category][i][0])
        S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
        music_spectograms[category].append(S_db)
```

ŒëœÇ ŒøœÄœÑŒπŒ∫ŒøœÄŒøŒπŒÆœÉŒøœÖŒºŒµ ŒºŒµœÅŒπŒ∫Œ¨ Œ±œÄœå Œ±œÖœÑŒ¨. ŒåœÄœâœÇ Œ∫Œ±Œπ ŒºŒµ œÑŒ± waveforms, œÑŒø œÄœÅœéœÑŒø Œ∫Œ¨Œ∏Œµ Œ∫Œ±œÑŒ∑Œ≥ŒøœÅŒØŒ±œÇ.

```python
import librosa.display

SAMPLE_TO_SHOW = 0 # The first sample

categories = ((CATEGORIES_NAMES[0], CATEGORIES_NAMES[1]), (CATEGORIES_NAMES[2], CATEGORIES_NAMES[3]))

fig, axes = plt.subplots(2, 2, figsize=(14, 8))

for i in range(len(categories)):
    for j in range(len(categories[i])):
        img = librosa.display.specshow(
            music_spectograms[categories[i][j]][SAMPLE_TO_SHOW], 
            x_axis='time', y_axis='log', 
            ax=axes[i][j], 
            cmap=custom_cmap(plt.get_cmap('magma'))
        )
        
        title = 'Category: ' + categories[i][j] + '\n Song: ' + music_audio[categories[i][j]][0][2].split('\\')[-1][:-4]
        axes[i][j].set_title(title, fontsize=10)
        axes[i][j].set_xlabel('Time (s)')
        axes[i][j].set_ylabel('Frequency (Hz)')
        
        fig.colorbar(img, ax=axes[i][j], format='%2.0f dB')

fig.suptitle('\nSpectrogram for every music category\n', fontsize=25)
plt.tight_layout()
plt.show()
```

## ŒïœÜŒ±œÅŒºŒøŒ≥ŒÆ œÑŒ∑œÇ Œ∫ŒªŒØŒºŒ±Œ∫Œ±œÇ Mel œÉœÑŒ± œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒÆŒºŒ±œÑŒ± üîäüëÇ

Œ£œÑŒ∑ œÉœÖŒΩŒ≠œáŒµŒπŒ± Œ∏Œ± __ŒºŒµœÑŒ±œÑœÅŒ≠œàŒøœÖŒºŒµ__ œÑŒ± œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒÆŒºŒ±œÑŒ¨ ŒºŒ±œÇ œÉŒµ __Œ∫Œ±œÑŒ¨ŒªŒªŒ∑ŒªŒ∑ ŒºŒøœÅœÜŒÆ__ Œ∑ ŒøœÄŒøŒØŒ± ŒºŒ±œÇ œÄœÅŒøœÉœÜŒ≠œÅŒµŒπ ŒºŒØŒ± œÄŒπŒø __Œ≤ŒøŒªŒπŒ∫ŒÆ__ Œ∫Œ±Œπ __Œ∫Œ±œÑŒ±ŒΩŒøŒ∑œÑŒÆ__ Œ¥ŒπŒ±œáŒµŒØœÅŒπœÉŒ∑ œÑœâŒΩ Œ¥ŒµŒ¥ŒøŒºŒ≠ŒΩœâŒΩ œÉœÑŒ∑ Œ¥ŒπŒ±Œ¥ŒπŒ∫Œ±œÉŒØŒ± ŒµŒ∫œÄŒ±ŒØŒ¥ŒµœÖœÉŒ∑œÇ œÑŒøœÖ ŒºŒøŒΩœÑŒ≠ŒªŒøœÖ. ŒëœÖœÑŒÆ Œ∑ ŒºŒøœÅœÜŒÆ ŒºŒøŒπŒ¨Œ∂ŒµŒπ ŒºŒµ œÑŒ± ŒºŒ≠œáœÅŒπ œÑœéœÅŒ± œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒÆŒºŒ±œÑŒ± œÄŒøœÖ Œ≠œáŒøœÖŒºŒµ, ŒºœåŒΩŒø œÄŒøœÖ Œ±œÖœÑŒÆ œÑŒ∑ œÜŒøœÅŒ¨ Œ∏Œ± ŒµœÜŒ±œÅŒºœåœÉŒøœÖŒºŒµ œÉœÑŒ∑ Œ∫ŒªŒØŒºŒ±Œ∫Œ± [__mel__][mel_link], ŒºŒØŒ± Œ∫ŒªŒØŒºŒ±Œ∫Œ± Œ∑ ŒøœÄŒøŒØŒ± ŒµŒØŒΩŒ±Œπ Œ∫ŒøŒΩœÑŒ¨ œÉœÑŒøŒΩ œÑœÅœåœÄŒø ŒºŒµ œÑŒøŒΩ ŒøœÄŒøŒØŒø Œø Œ¨ŒΩŒ∏œÅœâœÄŒøœÇ __Œ±ŒΩœÑŒπŒªŒ±ŒºŒ≤Œ¨ŒΩŒµœÑŒ±Œπ__ œÑŒ± __Œ∑œáŒ∑œÑŒπŒ∫Œ¨ œÉŒÆŒºŒ±œÑŒ±__ ŒºŒ≠œÉœâ œÑŒ∑œÇ Œ±Œ∫ŒøŒÆœÇ. ŒúŒµ Œ±œÖœÑœåŒΩ œÑŒøŒΩ œÑœÅœåœÄŒø Œ∏Œ± ŒªŒ¨Œ≤ŒøœÖŒºŒµ œÑŒ± ŒªŒµŒ≥œåŒºŒµŒΩŒ± __mel spectograms__ ŒÆ Œ±œÄŒªœéœÇ __melgrams__. 

[mel_link]: https://en.wikipedia.org/wiki/Mel_scale

```python
import librosa
import numpy as np

mel_spectrograms = {key: [] for key in CATEGORIES_NAMES}

for category in CATEGORIES_NAMES:
    for i in range(len(music_audio[category])):
        audio_samples, sampling_rate = music_audio[category][i][0], music_audio[category][i][1]
        S = librosa.feature.melspectrogram(y=audio_samples, sr=sampling_rate, n_mels=128)
        S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
        mel_spectrograms[category].append(S_db_mel)
```

## ŒîŒµŒπŒ≥ŒºŒ±œÑŒøŒªŒ∑œàŒØŒ± œÑœâŒΩ Œ¥ŒµŒ¥ŒøŒºŒ≠ŒΩœâŒΩ ‚úÇÔ∏è

Œ£œÑŒ∑ œÉœÖŒΩŒ≠œáŒµŒπŒ± Œ∏Œ± __œáœâœÅŒØœÉŒøœÖŒºŒµ__ Œ∫Œ¨Œ∏Œµ œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ¨œÜŒ∑ŒºŒ± œÄŒøœÖ Œ≠œáŒøœÖŒºŒµ œÉŒµ __œÖœÄŒø-œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒÆŒºŒ±œÑŒ±__ ŒºŒÆŒ∫ŒøœÖœÇ _ŒµŒΩœåœÇ Œ¥ŒµœÖœÑŒµœÅŒøŒªŒ≠œÄœÑŒøœÖ_. ŒúŒµ Œ±œÖœÑœåŒΩ œÑŒøŒΩ œÑœÅœåœÄŒø Œ∏Œ± Œ≠œáŒøœÖŒºŒµ œÉœÑŒ∑ŒΩ Œ∫Œ±œÑŒøœáŒÆ ŒºŒ±œÇ ŒºŒØŒ± œÄŒªŒ∑Œ∏œéœÅŒ± Œ¥ŒµŒ¥ŒøŒºŒ≠ŒΩœâŒΩ __ŒºŒπŒ∫œÅœéŒΩ Œ¥ŒπŒ±œÉœÑŒ¨œÉŒµœâŒΩ__ ŒºŒµ œÑŒ∑ŒΩ ŒøœÄŒøŒØŒ± Œ∏Œ± __ŒµŒ∫œÄŒ±ŒπŒ¥ŒµœçœÉŒøœÖŒºŒµ__ Œ∫Œ±Œπ Œ∏Œ± __Œ±ŒæŒπŒøŒªŒøŒ≥ŒÆœÉŒøœÖŒºŒµ__ œÑŒø ŒºŒøŒΩœÑŒ≠ŒªŒø ŒºŒ±œÇ.

Œ†Œ±œÅŒ±Œ∫Œ¨œÑœâ ŒøœÅŒØœÉŒøœÖŒºŒµ ŒºŒØŒ± œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑ Œ∑ ŒøœÄŒøŒØŒ± Œ¥Œ≠œáŒµœÑŒ±Œπ Œ≠ŒΩŒ± ŒµœÄŒπœÉœÑœÅŒ≠œÜŒµŒπ ŒºŒØŒ± ŒªŒØœÉœÑŒ± ŒºŒµ œÑŒ± œÖœÄŒø-œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒÆŒºŒ±œÑŒ± œÄŒøœÖ œÄœÅŒøŒ∫œçœÄœÑŒøœÖŒΩ, ŒµŒ¨ŒΩ œáœâœÅŒØœÉŒøœÖŒºŒµ œÑŒø Œ±œÅœáŒπŒ∫œå œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ¨œÜŒ∑ŒºŒ± œÉŒµ Œ∫ŒøŒºŒºŒ¨œÑŒπŒ± ŒºŒµŒ≥Œ≠Œ∏ŒøœÖœÇ $128 \times \text{chunks}$ œåœÄŒøœÖ œÑŒ± __chunks__ œÑŒ± ŒøœÅŒØœÉŒµŒπ Œø œáœÅŒÆœÉœÑŒ∑œÇ Œ∫Œ±Œπ œÑŒ± œÄŒµœÅŒΩŒ¨ŒµŒπ œâœÇ œåœÅŒπœÉŒºŒ± œÉœÑŒ∑ œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑.

```python
def split_spec_to_subspects(spectrogram, chunks) -> list:
    """ Splits the given spectrogram into a list of sub-spectrograms, 
        each one having a shape of 128 x chunks. """
    
    subspects = [] # Intialize the list of the sub-spectrograms
    
    # Loop through the main spectrogram and split it into individual segments of shape 128 x chunks
    for start in range(0, spectrogram.shape[1], chunks):
        end = min(start + chunks, spectrogram.shape[1])
        subspec = spectrogram[:, start:end]
        subspects.append(subspec)

    return subspects
```

ŒòŒ± œáœÅŒ∑œÉŒ∑ŒºŒøœÄŒøŒπŒÆœÉŒøœÖŒºŒµ œÑŒ∑ŒΩ œÄŒ±œÅŒ±œÄŒ¨ŒΩœâ œÉœÖŒΩŒ¨œÅœÑŒ∑œÉŒ∑ Œ≥ŒπŒ± ŒΩŒ± __'œáœâœÅŒØœÉŒøœÖŒºŒµ'__ œåŒªŒ± œÑŒ± œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒÆŒºŒ±œÑŒ± œÑŒ∑œÇ __Œ∫Œ¨Œ∏Œµ Œ∫Œ±œÑŒ∑Œ≥ŒøœÅŒØŒ±œÇ__, œÑŒ± ŒøœÄŒøŒØŒ± Œ∏Œ± Œ±œÄŒøŒ∏Œ∑Œ∫Œ≠œÖœÉŒøœÖŒºŒµ œÉŒµ ŒºŒØŒ± ŒºŒµŒ≥Œ¨ŒªŒ∑ __ŒªŒØœÉœÑŒ±__ Œ±œÄœå _œÖœÄŒø-œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒÆŒºŒ±œÑŒ±_. ŒòŒ± œÜœÅŒøŒΩœÑŒØœÉŒøœÖŒºŒµ ŒµœÄŒØœÉŒ∑œÇ ŒΩŒ± Œ±œÄŒøŒ∏Œ∑Œ∫ŒµœçœÉŒøœÖŒºŒµ Œ∫Œ±Œπ œÑŒ∑ŒΩ __Œ∫Œ±œÑŒ∑Œ≥ŒøœÅŒØŒ±__ œÑŒøœÖ Œ∫Œ¨Œ∏Œµ œÖœÄŒø-œÜŒ±œÉŒºŒ±œÑŒøŒ≥œÅŒ±œÜŒÆŒºŒ±œÑŒøœÇ œÉŒµ ŒºŒØŒ± Œ¥ŒµœçœÑŒµœÅŒ∑ ŒªŒØœÉœÑŒ± ŒØŒ¥ŒπŒøœÖ ŒºŒÆŒ∫ŒøœÖœÇ ŒºŒµ œÑŒ∑ œÄœÅœéœÑŒ∑. ŒàœÑœÉŒπ Œ∏Œ± Œ≠œáŒøœÖŒºŒµ

```python
# Initialize the lists to contain the sub-spectrograms and their category name
total_subspects, total_categories = [], []

# Apply the spliting process to every spectrogram of every category
for category in CATEGORIES_NAMES:
    for i in range(len(music_audio[category])):
        audio_spec = mel_spectrograms[category][i]
        
        # Get the sequence of sub-spectrograms of the current spectrogram
        subspects = split_spec_to_subspects(audio_spec, int(audio_spec.shape[1] / music_audio[category][i][3]))[:-1]
        
        # Store the subspectrogram and its category
        total_subspects.extend(subspects)
        total_categories.extend([category for _ in range(len(subspects))])
```
